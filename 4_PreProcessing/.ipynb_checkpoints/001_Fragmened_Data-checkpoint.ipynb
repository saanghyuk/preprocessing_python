```python
df1 = pd.read_csv("날짜포맷이다른데이터1.csv")
df2 = pd.read_csv("날짜포맷이다른데이터2.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
#df1의 날짜 타입을 df2의 날짜 타입으로 변경하는 함수 작성 및 적용
def date_type_converter(value):
    YYYY, MM, DD = value.split('-')
    return YYYY + "년 " + str(int(MM)) + "월 " + str(int(DD)) + "일"

df1['날짜'] = df1['날짜'].apply(date_type_converter)
```

```python
# 데이터 병합
merged_df = pd.merge(df1, df2, on = '날짜')
```

```python
merged_df.head()
```

```python
df3 = pd.read_excel("지하철역_위경도.xlsx")
```

```python
df3.head()
```

```python
# 거리 행렬 생성을 위한 컬럼 추출
df_location = df[['경도', '위도']]
df3_location = df3[['경도', '위도']]
```

```python
df_location.head()
```

```python
# !pip install haversine # 위경도 거리 계산을 위한 모듈 설치
```

지구상에서 두 지점간의 거리를 구할때는, 유클리디안 거리나 맨하탄 거리를 쓰는게 적절하지는 않음. harversine이라는 척도를 쓰면 된다. 

```python
# 거리 행렬 생성
from scipy.spatial.distance import cdist
from haversine import haversine

dist_mat = cdist(df_location, df3_location, metric = haversine)
# 'euclidean', 'cityblock', 'jaccard'
```

```python
dist_mat

# 행은 각각 row가 거래 매물들
# 열은 각각이 지하철역
```

```python
dist_mat.argsort()
```

```python
dist_mat.shape
```

```python
dist_mat.argsort()[:, 0]
```

```python
close_subway_index1 = dist_mat.argsort(axis=1)[:, 0] # axis = 0
close_subway_index2 = dist_mat.argsort()[:, 0]
close_subway_index3 = dist_mat.argsort(axis=0)[:, 0]
close_subway_index4 = dist_mat.argsort(axis=-1)[:, 0]
close_subway_index1.shape
```

```python
(close_subway_index1==close_subway_index4).sum()
```

```python
(close_subway_index1==close_subway_index3).sum()
```

```python
close_subway_index = dist_mat.argsort(axis=1)[:, 0] # axis = 0
df['가까운역'] = df3.iloc[close_subway_index]['역명'].values 
# 새로운 시리즈를 만들 때는 list, ndarray를 사용하는 것이 바람직하다!

df['가까운역까지_거리'] = dist_mat[close_subway_index][:, 0]
```
