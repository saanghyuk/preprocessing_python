## Line Chart

#### pyplot.plot
- x, y : x, y 축에 들어갈 값(iterable한 객체, x[i], y[i]가 하나의 점이 되므로 길이가 같아야 함)
- linewidth : 선 두께
- marker : 마커 종류
- markersize : 마커 크기
- color : 선 색상
- linestyle : 선 스타일
- label : 범례

#### DataFrame.plot()
- kind: 그래프 종류(line이면 선 그래프)
- x : x축에 들어갈 컬럼명(입력하지 않으면, index가 x축)
- y: y축에 들어갈 컬럼명(목록)
- xtics, ytics도 설정 가능
- 단 이것으로는 단순한 그래프만 그리고, 더 디테일한 그래프를 그리려면 pyplot을 사용할 것

## Scatter Plot

#### DataFrame.plot()

- 단 이것으로는 단순한 그래프만 그리고, 더 디테일한 그래프를 그리려면 pyplot을 사용할 것

```python
# 년도 - 분기를 잇는 새로운 변수 생성
df['분기'] = df['년도(year)'].astype(str) + '-' + df['분기(4)'].astype(str)
df['분기'].head()
```

```python
grouped_df = df.groupby(['분기', '대리점'], as_index = False)['수량'].sum()
grouped_df
```

#### pyplot을 이용한 그래프 그리기

```python
grouped_df['분기'].unique()
```

```python
grouped_df.loc[cond]['분기']
```

```python
# 그래프 기본 설정
plt.title("날짜 및 대리점별 출하량")
plt.xlabel("분기")
plt.ylabel("출하량 (개수)")

xtick_range = range(len(grouped_df['분기'].unique()))
xtick_label = grouped_df['분기'].unique()
plt.xticks(xtick_range, xtick_label)

for agent in grouped_df['대리점'].unique():
    cond = grouped_df['대리점'] == agent
    plt.scatter(grouped_df.loc[cond]['분기'], grouped_df.loc[cond]['수량'], s = 100, marker = '*')
```

#### DataFrame을 이용하여 직접 그래프 그리기

```python
grouped_df.head() 
# 대리점별로 데이터를 나누거나, pivot을 해야만 df.plot을 이용하기 적합해짐
```

```python
pivot_df = pd.pivot_table(data = grouped_df, values = '수량', index = '분기', columns = '대리점', aggfunc = 'sum')
pivot_df = pivot_df.add_suffix("_출하량") # 대리점명이 전부 컬럼으로 되기 때문에, suffix를 추가시킴

pivot_df
```

```python
pivot_df['분기'] = list(range(len(pivot_df))) # x축에 쓰기 위해 임의의 변수를 생성함
pivot_df['분기'].head()
```

```python
pivot_df.columns
```

```python
# 그래프 기본 설정
# x="분기"는, 숫자로 안넣으면 x축 위치가 잘 안잡히는 경우가 있기에 숫자로 넣고, 추후에 xtick으로 다시 조정. 
pivot_df.plot(kind = 'scatter', x = '분기', y = '대리점 1_출하량')
plt.title("날짜별 상품 판매량")
plt.xlabel("분기")
plt.ylabel("출하량 (개수)")

xtick_range = range(len(grouped_df['분기'].unique()))
xtick_label = grouped_df['분기'].unique()

plt.xticks(xtick_range, xtick_label)
```

## Pie Chart

#### pyplot.pie 함수를 사용하여 파이 차트를 그린다. 
- x :각 Pie의 크기
- labels : 각 pie에 부탁되는 라벨 
- labeldistance : 라벨 간 거리(원의 중심과 라벨의 거리 의미. 보통 0.3~0.4정도 쓴다)
- normalize : 비율을 나타낼 것인지 여부
- **autopct** : 위에 표시될 글자 형태 (예, 소숫점 첫자리 까지 "%1.1f%%", 정수형 "%1d%%")
- color : 배열로 설정해서 각 파트의 색상을 설정 가능

#### DataFrame.plot()
- kind: 그래프 종류(pie 이면 파이 그래프)
- x : 각 pie의 크기
- y: 각 pie에 부탁되는 라벨

```python
import pandas as pd
import os
os.chdir(r"/Users/sanghyuk/Documents/preprocessing_python/lecture_source/1. 데이터 핸들링/데이터")
```

```python
df = pd.read_csv("판매실적통합본.csv", engine = "python", encoding='CP949')
df.head()
```

```python
# 제품군별 수량합계 계산
grouped_df = df.groupby(['제품군'], as_index = False)['수량'].sum()
```

```python
grouped_df
```

```python
grouped_df.sort_values(by = '수량')['수량']
```

```python
# 기타 행 추가
other_row = pd.DataFrame({"제품군":['기타'], 
                          '수량':[grouped_df.sort_values(by = '수량')['수량'].iloc[:4].sum()]})

grouped_df = pd.concat([grouped_df, other_row], axis = 0, ignore_index = True)
grouped_df = grouped_df.sort_values(by = '수량', ascending = False).iloc[:3]
grouped_df
```

```python
print(other_row)
```

```python
# 그래프 기본 설정
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.size"] = 15
plt.rcParams["figure.figsize"] = (10, 8)
```

```python
# 제품군별 출하량 합계 계산
plt.title("제품군별 출하량 비율")

plt.pie(x = grouped_df['수량'],
        labels = grouped_df['제품군'],
       labeldistance = 0.3,
       autopct = '%1.1f%%')
```

## Box Plot

#### pyplot.pie 함수를 사용하여 박스 플롯을 그린다. 
- x : boxplot을 그리기 위한 데이터(2차원인 경우, 여러 개의 박스 플롯이 그려진다. 즉, 하나의 열이 하나의 박스 플롯으로 생성됨)


#### DataFrame.boxplot()
- column: box plot을 그릴 컬럼 목록

```python
import pandas as pd
import os
os.chdir(r"/Users/sanghyuk/Documents/preprocessing_python/lecture_source/1. 데이터 핸들링/데이터")
```

```python
df = pd.read_csv("온라인_판매기록.csv", engine = "python", encoding='CP949')
df.head()
```

```python
# pyplot을 이용한 boxplot 그리기
X = df.groupby(['쇼핑몰 유형'])['판매금액'].apply(list) # 쇼핑몰 유형에 따른 판매 금액 목록을 리스트화
X
```

```python
# box plot은 starting index가 1
np.arange(1, len(X)+1)
```

```python
plt.title("쇼핑몰 유형에 따른 판매금액 분포")
plt.boxplot(X.values)
plt.xticks(np.arange(1, len(X)+1), X.index)
```
