GroupBy는 조건부 통계량을 계산하기 위한 방법
- `df.groupby(분할기준 column)[적용기준 column].집계함수`
- as_index : 분할 기준 컬럼들을 인덱스로 사용할 것인지 여부(default:True)
- pivot_table과 group_by 모두 결과로 주는 값은 같음. 단, **뱉어주는 데이터의 형태/구조가 차이**가 있을 뿐임.

```python
df = pd.read_csv("온라인_판매기록.csv", engine = "python", encoding='CP949')
```

```python
df.groupby(['쇼핑몰 유형'])['수량'].mean()
```

```python
df.groupby(['쇼핑몰 유형'], as_index = False)['수량'].mean()
```

```python
df.groupby(['쇼핑몰 유형'])['수량', '판매금액'].agg(['mean', 'max'])
```

```python
# value parameter에 리스트/np.array/series가 통으로 다 들어온다고 생각하면 됨. 
def my_func(value):
    return max(value) - min(value)


# Multi Index Case
df.groupby(['쇼핑몰 유형', '제품'])['수량', '판매금액'].agg(my_func)
```

```python
df.groupby(['쇼핑몰 유형', '제품'], as_index=False)['수량', '판매금액'].agg(my_func)
```
